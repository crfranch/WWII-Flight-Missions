```python
import pandas as pd
```

```python
# read data
missions = pd.read_csv("wwii_flight_missions.csv")

missions.head()
```

```python
# get a list of the column names
list(missions)
```

```python
# get unique theaters
missions.THEATER.unique()
```

```python
# get unique target countries
missions.TGT_COUNTRY.unique()
```

```python
# see how many times each country was targeted
tgtCountry_count = missions.groupby('TGT_COUNTRY').count()['MSNDATE']

# make a dataframe
tgtCountry_count_df = pd.DataFrame(tgtCountry_count)

tgtCountry_count_df.head()
```

```python
# sort target country missions
tgtCountry_count_df = tgtCountry_count_df.sort_values(by = 'MSNDATE', ascending = False)

tgtCountry_count_df.head(10)
```

```python
# total # of flight missions
total_missions = len(missions)

print(total_missions)
```

```python
# see how many times each location was targeted
tgtLocation_count = missions.groupby('TGT_LOCATION').count()['MSNDATE']

# make a dataframe
tgtLocation_count_df = pd.DataFrame(tgtLocation_count)

tgtLocation_count_df.head()
```

```python
# sort target locations
tgtLocation_count_df = tgtLocation_count_df.sort_values(by = 'MSNDATE', ascending = False)

tgtLocation_count_df.head(10)
```

```python
# get unique aircraft name
missions.AIRCRAFT_NAME.unique()
```

```python
# count aircraft types used in missions
aircraft_count = missions.groupby('AIRCRAFT_NAME').count()['MSNDATE']

# make a dataframe
aircraft_count_df = pd.DataFrame(aircraft_count)

aircraft_count_df.head()
```

```python
# sort aircraft missions
aircraft_count_df = aircraft_count_df.sort_values(by = 'MSNDATE', ascending = False)

aircraft_count_df.head(10)
```

```python
# unique mission types
missions.MSN_TYPE.unique()
```

```python
# count callsigns 
callSign_count = missions.groupby('CALLSIGN').count()['MSNDATE']

# make a dataframe
callSign_count_df = pd.DataFrame(callSign_count)

callSign_count_df.head()
```

```python
# sort callsigns
callSign_count_df = callSign_count_df.sort_values(by = 'MSNDATE', ascending = False)

callSign_count_df.head(10)
```

```python
# parse year from mission dates and make a new column with just the year
missions['MSNDATE'] = pd.to_datetime(missions['MSNDATE'])

missions['YEAR'] = missions['MSNDATE'].dt.year

missions.head()
```

```python
# change year from int to string
missions = missions.astype({"YEAR": str})
```

```python
# find the number missions flown per year
missions_per_year = missions.groupby('YEAR').count()['MSNDATE']

# make a dataframe
missions_per_year_df = pd.DataFrame(missions_per_year)

missions_per_year_df
```

```python
# drop unneccessary columns
missions_clean = missions.drop(['SPARES_RETURN_AC','WX_FAIL_AC','MECH_FAIL_AC','MISC_FAIL_AC','TARGET_COMMENT','MISSION_COMMENTS','SOURCE','DATABASE_EDIT_COMMENTS', 'AC_LOST','AC_DAMAGED','AC_AIRBORNE','AC_DROPPING','TIME_OVER_TARGET','SIGHTING_METHOD_CODE','SIGHTING_EXPLANATION'], axis = 1)
```
